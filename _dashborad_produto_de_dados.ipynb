**Média Móvel**

A média móvel é um indicador de faz parte da Análise Técnica de ações. Com ela, é possivel indentificar o equilíbrio dos preços no mercado, observando tendências de alta, neutra ou baixa. A representação gráfica das médidas móveis é normalmente feita por uma linha, que se movimenta conforme os dados novos recebidos para o cálculo.

**O que é a média móvel?**

A média móvel é um indicador importantew de suaviza os dados dos preçoes das ações para formar um indicador que segue a tendência.

**E como fazer esse cálculo?**

Basta utilizar a fórmula: MMA = Soma dos preços de fechamento + Número de dias


Considere a média móvel aritmética de 5 dias e sua evolução ao longo de 3 dias, em que os preços diários de fechamento são: 11, 12, 13, 14, 15, 16 e 17.

Primeiro dia da MMa de 5 dias: (11 + 12 + 13 + 14 + 15)/ 5 = 13
Segundo dia da MMA de 5 dias: (12  + 13 + 14 + 15 + 16)/ 5 = 14
Terceiro dia da MMA de 5 dias: (13 + 14 + 15 + 16 + 17)/ 5 = 15

```python
# Media Simples
Lista = [11, 12, 13, 15, 16, 17]

from statistics import mean as Media
print(f'Media: { Media(Lista) }')
print(f'Media: { (11+12+13+15+16+17) / len(Lista)}')
```

```python
# Periodo da média móvel
n = 5

for Intervalo in range(3):
  print(f'{Intervalo} Dia: {sum(Lista[Intervalo:n]) / 5}')
```

**Função rolling**

A função rolling() é uma função do pandas que permite calcular estatísticas em janelas deslizantes de dados.
Ele é muito útil para calcular médias móveis, desvios padrão, somatórios, mínimos e máximos, entre outras estatísticas.

```python
Periodo = 5
Dados ['Close'].rolling( window=Periodo).mean()
```

```python
#Plot

# Estilos
plt.style.use('seaborn-darkgrid')

#Tamanho
plt.figure( figsize=(12, 6) )

#Plots
plt.plot( Dados['Close'] )
plt.plot( Dados ['Close'].rolling( window=Periodo).mean() )
plt.plot( Dados ['Close'].rolling( window=Periodo + 25 ).mean() )

# Labels
plt.ylabel('Valor de Fachamento')
plt.xlabel('Periodo')

#Legenda
plt.legend(['Close', 'MM 5 dias', 'MM 25 dias'])
plt.title('Analise preço de fechamento');
```

```python
# Cria um gráfico de área (área preenchida)
Area = px.area(
    data_frame=Dados.reset_index(),
    y='Close', x='Date',
    title=f'{Simbolo}'
)

# Atualiza as configurações dos eixos x (datas) do gráfico
Area.update_xaxes(
    title_text='Data',
    rangeslider_visible=True,
    rangeselector=dict(
        buttons=list([
            dict(count=1, label='1M', step='month', stepmode='backward'),
            dict(count=6, label='6M', step='month', stepmode='backward'),
            dict(count=1, label='1Y', step='year', stepmode='backward'),
            dict(step='all')
        ])
    )
)

# Ajustando o eixo Y
Area.update_yaxes( title_text='Price in USD', ticksuffix='$')

# Ajustar o titulo
Area.update_layout(
    showlegend=True,
    title={
        'text':f'{Simbolo} ({Dados.index.min().year}-Presente)',
        'y': 0.9,
        'x': 0.5,
        'xanchor': 'center',
        'yanchor': 'top'
      },
      template="plotly_white"
    )

Area
```

**Candlestick Charts**

É um estilo de gráfico financeiro que descreve abertura, alta, baixa e fehchamento par auma determinada xcoordenada(tempo mais provável). As caixas respresentam a dispersão entre osvalores opene cloesee as linhas representam a despersão entre os valores lowe high. Pontos de amostragem onde o valor de fechamento é maior (inferior) do que o valor de abertura são chamados de crescentes (decrescentes). Por padrão, velas crescentes são desenhadas em verde, enquanto as descrescentes são desenhadas em vermelho.

```python
# Grafico clandestik
Grafico_Candlestick = go.Figure(
    data=[
        go.Candlestick(
            x=Dados.index,
            open=Dados['Open'],
            high=Dados['High'],
            low=Dados['Low'],
            close=Dados['Close'],
            increasing_line_color='cyan',
            decreasing_line_color='gray'

        )
    ]
)

Grafico_Candlestick.update_layout(
    xaxis_rangeslider_visible=False,
    title='Análise Fechamento',
    xaxis_title='Periodo',
    yaxis_title='Preço de Fechamento'
)

Grafico_Candlestick.show()
```

```python
# Importando a Imagem
from PIL import Image
Logo_Bitcoin = Image.open("/content/icone_bitcoin.png")
Logo_Bitcoin
```

```python
# COntrução do Relatório
Relatorio = make_subplots(
    rows=2,
    cols=1,
    specs=[
        [ {'type':'scatter'} ],
        [ {'type':'scatter'} ],
    ],
    subplot_titles=('Cotação', 'Fechamento'),
    # Compartilhar mesmo eixo X
    shared_xaxes=True,
    # Largura da vertical
    vertical_spacing=0.075
)

Relatorio.update_layout(
    #Dimensão
    width=1000, height=800,

    #Titulo no relatório
    title_text='<b>Data Analytics</b> <br>Cotação do Bitcoin</b>',

    #Estilo
    template='plotly_dark',

    #Ajuste de layout gráficos
    paper_bgcolor='rgb(0, 0, 0)',
    plot_bgcolor='rgb(10, 10, 10)',

    # Fonte do texto
    font=dict(color='white')
)

#Colocando um grafico
Relatorio.add_trace(
    go.Candlestick(
        x=Dados.index,
        open=Dados['Open'],
        high=Dados['High'],
        low=Dados['Low'],
        close=Dados['Close'],
        increasing_line_color='cyan',
        decreasing_line_color='gray'
    ), row=1, col=1
)

# Retirando a barra visivel
Relatorio.update_layout(
    xaxis_rangeslider_visible=False
)

# Adicionando um gráfico na 2° Posição
Relatorio.add_trace(
    go.Scatter(
        x=Dados.index,
        y=Dados['Close'],
        mode='lines',
        name='Fechamento',
        line=dict(color='green')
    ),
    row=2, col=1
)

Relatorio.add_trace(
    go.Scatter(
        x=Dados.index,
        y=Dados['Media_Movel'],
        mode='lines',
        name='Média Móvel',
        line=dict(color='rgb(158, 58, 171)')
    ),
    row=2, col=1
)

# Ajustes
Relatorio.update_layout(
    legend=dict(
        # Posição
        orientation='h',
        # Eixo X e Y
        yanchor='bottom',
        y=1.02,
        xanchor='right',
        x=1,
        font=dict( size=9)
    )

)

# Acionar texto
Relatorio.add_annotation(
    # Seta (desativado)
    showarrow=False,
    #texto
    text='Análise Para apurar cotação do BITCOIN',
    #Tamanho
    font=dict(size=12),

    # Aponta as direções do texto
    xref='x domain',
    x=1,
    yref='y domain',
    y=-1.4
)

Relatorio.add_annotation(
    showarrow=False,
    text='By: Denilson Delgado C. |',
    font=dict(size=12),

    # Aponta as direções do texto
    xref='x domain',
    x=0,
    yref='y domain',
    y=-1.4
)

Relatorio
```
